```python
# Loading dataset

import pandas as pd
lkd_data = pd.read_csv('/content/Connections_LinkedIn.csv')
```

```python
lkd_data.drop(['Email Address'],axis=1,inplace=True)
```

```python
lkd_data.head(20)
```

# Connectivity Over Time

```python
lkd_new = lkd_data.sort_values(by='Connected On')

import plotly.express as px

df=lkd_new.groupby(by='Connected On').count().reset_index()
px.line(df,
        x = 'Connected On',
        y = 'First Name',
        labels = {'First Name':'Number of Connections'},
        title = 'Connectivity Over Time')
```

# Companies Of the Connections

```python
lkd_new['Company']
```

```python
company_variety = lkd_new.groupby(by='Company').count().reset_index()
company_variety
```

```python
# descending order
new_Com_variety=company_variety.sort_values(by='Connected On',ascending=False).reset_index(drop=True)
```

```python
new_Com_variety
```

```python
# Visualization

fig = px.bar(company_variety[:200],
       x = 'Company',
       y = 'Connected On',
       labels = {'Connected On':'Number of Connections'},
       width = 1200,
       height = 900,
       title = 'Companies Connections Work At')
fig
```

```python
 # Tree Plot for same data
fig2 = px.treemap(company_variety[:200],path=['Company','Position'], 
       values = 'Connected On',
       labels = {'Connected On':'Number of Connections'},
       width = 1500,
       height = 850,
       title = 'Companies Connections Work At')
fig2
```

# WordCloud Visualization

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

def CreateWordCloud(text):
  wordcloud = WordCloud(width=1000,height=900,
                        background_color='black',
                        min_font_size= 10,
                        colormap = 'Set2').generate(text)

  fig5 = plt.figure(figsize=(15,10))
  plt.imshow(wordcloud,interpolation = 'bilinear')

  plt.show()
  return fig5
```

```python
positions = ' '.join(lkd_new[~lkd_new.Position.isnull()].Position.unique())
```

```python
CreateWordCloud(positions)
```
